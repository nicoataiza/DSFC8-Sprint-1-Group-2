# Mapping

Loading shapefile

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
pd.set_option('max_columns', 100)
```

```python
shapefile = gpd.read_file('data/Provinces/Provinces.shp')
shapefile.head(5)
```

```python
type(shapefile)
```

```python
shapefile["x"] = shapefile.geometry.centroid.x
shapefile["y"] = shapefile.geometry.centroid.y

shapefile
```

```python
shapefile.plot()
```

Loading the data

```python
crime = pd.read_csv('data/crime_rate.csv')
```

```python
crime.head()
```

```python
[x for x in shapefile["REGION"].unique() if x.upper() not in crime["Region"].unique()]
```

```python
region = {}

for i in crime["Region"].unique(): 
    s_region = [x for x in shapefile["REGION"].unique() if i == x.upper()]
    if len(s_region) == 1:
        region[i] = s_region[0]
    else:
        region[i] = 'INPUT'

region
```

```python
region['NCR'] = 'Metropolitan Manila'
region['REGION I'] = 'Ilocos Region (Region I)'
region['REGION II'] = 'Cagayan Valley (Region II)'
region['REGION III'] = 'Central Luzon (Region III)'
region['REGION IV-A'] = 'CALABARZON (Region IV-A)'
region['REGION IV-B'] = 'MIMAROPA (Region IV-B)'
region['REGION V'] = 'Bicol Region (Region V)'
region['REGION VI'] = 'Western Visayas (Region VI)'
region['REGION VII'] = 'Central Visayas (Region VII)'
region['REGION VIII'] = 'Eastern Visayas (Region VIII)'
region['REGION IX'] = 'Zamboanga Peninsula (Region IX)'
region['REGION X'] = 'Northern Mindanao (Region X)'
region['REGION XI'] = 'Davao Region (Region XI)'
region['REGION XII'] = 'SOCCSKSARGEN (Region XII)'
region['ARMM'] = 'Autonomous Region of Muslim Mindanao (ARMM)'
region['CAR'] = 'Cordillera Administrative Region (CAR)'
region['CARAGA'] = 'Caraga (Region XIII)'
```

```python
# Replace province name
crime["Region"] = crime["Region"].replace(region)
print(sorted(crime["Region"].unique()))
```

```python
crime.set_index('Region', inplace = True)
```

```python
crime.head()
```

```python
# Complete the code

merged_data = pd.merge(shapefile, crime, left_on='REGION', right_on='Region')
merged_data.head(5)
```

```python
type(merged_data)
```

```python
#for voter
voter = pd.read_csv('data/2016-2019-voter-data.csv')
```

```python
province = {}

for i in voter["Province"].unique(): 
    s_province = [x for x in shapefile["PROVINCE"].unique() if i == x.upper()]
    if len(s_province) == 1:
        province[i] = s_province[0]
    else:
        province[i] = 'INPUT'
        
province
```

Once you have the final dataframe, plotting in geopandas is easy. You simply need to add arguments to change the colors.

```python
#set values
province['NCR'] = 'Metropolitan Manila'
province['DAVAO OCCIDENTAL'] = 'Shariff Kabunsuan'

# Replace province name
voter["Province"] = voter["Province"].replace(province)
print(sorted(voter["Province"].unique()))
```

```python
voter["Region"] = voter["Region"].replace(region)
print(sorted(voter["Region"].unique()))
```

```python
province_data = voter.groupby("Province")['2016-Registered_Literacy'].mean()

province_data.head(5)
```

```python
region_data = voter.groupby("Region")['2016-Registered_Voters'].sum()
region_data
```

```python
#merge shapefile and province_data

merged_data_p = pd.merge(shapefile, province_data, left_on='PROVINCE', right_on='Province')
merged_data_p.head(5)
```

```python
#merge shapefile and region_data

merged_data_r = pd.merge(shapefile, region_data, left_on='REGION', right_on='Region')
merged_data_r.head(5)
```

```python
merged_data_r['2016-Registered_Voters'] = merged_data_r['2016-Registered_Voters']/1000000
```

```python
# Complete the code
# set a variable that will call whatever column we want to visualise on the map
variable1 = "Rate per 100K"
variable2 = "2016-Registered_Voters"
# set the range for the choropleth
vmin1, vmax1 = merged_data["Rate per 100K"].min(), merged_data["Rate per 100K"].max()
vmin2, vmax2 = merged_data_r["2016-Registered_Voters"].min(), merged_data_r["2016-Registered_Voters"].max()
# create figure and axes for Matplotlib
fig, axes = plt.subplots(1,2, figsize=(15, 10))

axes[0].set_title("Index Crime Rate Total Population (in 100 Thousands)", size = 18)
merged_data.plot(column=variable1, cmap='OrRd', linewidth=0.8, ax=axes[0], edgecolor='0.8', vmin=vmin1, vmax=vmax1)
sm1 = plt.cm.ScalarMappable(cmap='OrRd', norm=plt.Normalize(vmin=vmin1, vmax=vmax1))
cbar = fig.colorbar(sm1,ax=axes[0] )

axes[1].set_title("2016 Registered Voters (in Millions)", size = 18)
merged_data_r.plot(column=variable2, cmap='OrRd', linewidth=0.8, ax=axes[1], edgecolor='0.8', vmin=vmin2, vmax=vmax2)
sm2 = plt.cm.ScalarMappable(cmap='OrRd', norm=plt.Normalize(vmin=vmin2, vmax=vmax2))
cbar = fig.colorbar(sm2, ax=axes[1])

plt.savefig("index crime per 100k Total Population 1.png", transparent=True, bbox_inches='tight')
plt.savefig("2016-Registered_Voters", transparent=True, bbox_inches='tight')
```
